```python
def Random_Sampling(file):
    df = pd.read_excel(file, header=5)
    randomlist = random.sample(range(0, len(df)), len(df))
    df['RandomNums'] = randomlist
    df_sorted = df.sort_values(by=['RandomNums'])
    df_threehundredonly = df_sorted.iloc[0:300]
    concord_listwithtags = df_threehundredonly['Concordance'].tolist()
    pattern = r'[A-Z\/_$\d]'
    concord_notags = []
    for line in concord_listwithtags:
        newstring = re.sub(pattern, '', line)
        concord_notags.append(newstring)
    df_threehundredonly.insert(2, 'NoTags', concord_notags)
    df_threehundredonly.to_excel(str(file) + '_Sampled.xlsx')
```
