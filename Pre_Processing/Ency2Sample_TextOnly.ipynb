```python
dfcon_only = pd.DataFrame(df,columns=['N', 'Concordance'])
```

```python
dfcon_only
```

```python
concord_listwithtags = dfcon_only['Concordance'].tolist()
```

```python
print(concord_listwithtags[0:20])
```

```python
print(len(concord_listwithtags))
```

```python
# remove all _ / Capital letters
import re
pattern = r'[A-Z\/_$\d]'
concord_notags = []

for line in concord_listwithtags:
    newstring = re.sub(pattern, '', line)
    concord_notags.append(newstring)
```

```python
print(concord_notags[0:20])
```

```python
print(len(concord_notags))
```

```python
dfcon_only['NoTags'] = concord_notags
dfcon_only
```

```python
dfcon_only.to_excel('Ency2Sample_WithoutTags.xlsx')
```
