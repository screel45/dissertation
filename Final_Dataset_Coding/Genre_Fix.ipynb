```python
# for each genre file, make a dictionary where concordance line is key and file number is value
zip(Ency_df.Concordance, Ency_df.File)
```

```python
Ency_dict = dict(zip(Ency_df.Concordance, Ency_df.File))
Med_dict = dict(zip(Med_df.Concordance, Med_df.File))
Milt_dict = dict(zip(Milt_df.Concordance, Milt_df.File))
News_dict = dict(zip(News_df.Concordance, News_df.File))
IT_dict = dict(zip(IT_df.Concordance, IT_df.File))
Law_dict = dict(zip(Law_df.Concordance, Law_df.File))
```

```python
print(len(Ency_dict))
print(len(Med_dict))
print(len(Milt_dict))
print(len(News_dict))
print(len(Law_dict))
print(len(IT_dict))
```

```python
# for each concordance line in master file, find corresponding dictionary key and add a column with dictionary value/file number
Ency_med = {**Ency_dict, **Med_dict}
Ency_med_milt = {**Ency_med, **Milt_dict}
Ency_med_milt_news = {**Ency_med_milt, **News_dict}
E_M_M_N_I = {**Ency_med_milt_news, **IT_dict}
All_dict = {**E_M_M_N_I, **Law_dict}
```

```python
print(len(All_dict))
```

```python
L1Arabicdf['Genre_File_Full'] = L1Arabicdf['Concordance'].map(All_dict)
L1Arabicdf
```
