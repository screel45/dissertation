```python
# Read in Z Arabic & Z English
# Pull off file column as list
# Run regex on file names and append to new list
# Attach new list as column next to File column

ZArabic_df = pd.read_excel("/Users/samanthacreel/Documents/Dissertation/Final_Datasets/FINAL/ZAEBUCArabic_FinalDataset.xlsx")
ZEnglish_df = pd.read_excel("/Users/samanthacreel/Documents/Dissertation/Final_Datasets/FINAL/ZAEBUCEnglish_FinalDataset.xlsx")
ZArabic_df
```

```python
Arabic_filenames = ZArabic_df['File'].tolist()
English_filenames = ZEnglish_df['File'].tolist()
print(len(Arabic_filenames))
print(Arabic_filenames[0:5])
print(len(English_filenames))
print(English_filenames[0:5])
```

```python
print(type(Arabic_filenames[2]))
for file in Arabic_filenames:
    if type(file) != 'str':
        print(file)
```

```python
ARShort_filenames = []
for file in Arabic_filenames:
    newfile = str(file)
    newnum = newfile[0:13]
    ARShort_filenames.append(newnum)
print(ARShort_filenames[0:5])
print(len(ARShort_filenames))
```

```python
AR_nums = []
for item in ARShort_filenames:
    if '.' in item:
        AR_nums.append(item[7:12])
    else:
        AR_nums.append(item[7:13])
print(AR_nums[0:5])
print(len(AR_nums))
```

```python
ENShort_filenames = []
for file in English_filenames:
    newfile = str(file)
    newnum = newfile[0:13]
    ENShort_filenames.append(newnum)
print(ENShort_filenames[0:5])
print(len(ENShort_filenames))
```

```python
EN_nums = []
for item in ENShort_filenames:
    if '.' in item:
        EN_nums.append(item[7:12])
    else:
        EN_nums.append(item[7:13])
print(EN_nums[0:5])
print(len(EN_nums))
```
