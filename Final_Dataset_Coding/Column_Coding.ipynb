```python
#If entry in GT column contains V+P, new column entry = BLANK
L1Arabicdf.GT.unique()
#Cases
# V + P = NaN
# ADV = ADV
# NP(S) = NP_S
# NP(O) = NP_O
# NP_IO - MANUAL
# NP_RES - MANUAL
# NP_RES_NP_S - MANUAL
# NP_O_ADV - MANUAL
```

```python
def L1Arabic_IMcolumn(df):
    if L1Arabicdf['GT'] == 'V + P':
        return 'None''
    if L1Arabicdf['GT'] == 'V + ADV + P':
        return 'ADV'
    if L1Arabicdf['GT'] == 'V + NP(S) + P':
        return NP_S
    if L1Arabicdf['GT'] == 'V + NP(O) + P':
        return NP_O
    if L1Arabicdf['GT'] == 'V + REP + P':
        return 'NP_RES'
    if L1Arabicdf['GT'] == 'V + REP + NP(S) + P':
        return 'NP_RES_NP_S'
    if L1Arabicdf['GT'] == 'V + NP(O) + ADV + P':
        return 'NP_O_ADV'
    if L1Arabicdf['GT'] == 'V + NP(IO) + P':
        return 'NP_IO'
```

```python
L1Arabicdf['Int_Mat'] = 0
L1Arabicdf
```

```python
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + P', 'Int_Mat'] = 'None' 
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + ADV + P', 'Int_Mat'] = 'ADV'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + NP(S) + P', 'Int_Mat'] = 'NP_S'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + NP(O) + P', 'Int_Mat'] = 'NP_O'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + REP + P', 'Int_Mat'] = 'NP_RES'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + REP + NP(S) + P', 'Int_Mat'] = 'NP_RES_NP_S'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + NP(O) + ADV + P', 'Int_Mat'] = 'NP_O_ADV'
L1Arabicdf.loc[L1Arabicdf['GT'] == 'V + NP(IO) + P', 'Int_Mat'] = 'NP_IO'
L1Arabicdf
```
