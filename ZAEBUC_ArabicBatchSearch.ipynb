```python
master_df["lemma_prep"] = master_df["Verb Lemma"] + " " + master_df["Preposition"]
```

```python
lemmas_andprepositions = master_df['lemma_prep'].unique().tolist()
print(lemmas_andprepositions[0:10])
print(len(lemmas_andprepositions))
```

```python
cleaning_combos = []
for combo in original_VPCs:
    combo = combo.replace(")","")
    cleaning_combos.append(combo)
cleaning_combos[0:10]
```

```python
clean_originalcombos = []
for combo in cleaning_combos:
    combo = combo.replace("(","")
    clean_originalcombos.append(combo)
clean_originalcombos[0:10]
```

```python
1059+1565
```

```python
all_combos = lemmas_andprepositions + clean_originalcombos
print(len(all_combos))
# all combos is the originals + preps plus the lemmas + preps
```

```python
ilas_r = []

for combo in all_combos:
    if "إلى" in combo:
        ilas_r.append(combo)
```

```python
print(len(ilas_r))
```

```python
ilas_w = []
for combo in all_combos:
    if "الى" in combo:
        ilas_w.append(combo)
print(ilas_w[0:10])
print(len(ilas_w))
```

```python
ilas_r_w = ilas_r + ilas_w
print(len(ilas_r_w))
```

```python
# taking the verbs w/ correctly spelled ila and copying them with the wrong spelling

separator = " "
all_ilas_r = []
for ila in ilas_r:
    ila = ila.split(separator, 1)[0] + " " + "الى"
    all_ilas_r.append(ila)
all_ilas_r[0:10]
```

```python
# taking the verbs with wrong spelling and copying them with right spelling
all_ilas_w = []
for ila in ilas_w:
    ila = ila.split(separator, 1)[0] + " " + "إلى"
    all_ilas_w.append(ila)
all_ilas_w[0:10]
```

```python
all_ilas_final = ilas_r_w + all_ilas_r + all_ilas_w
print(len(all_ilas_final))

# all ilas final has all the verbs and lemmas that appear with ila (w/ hamza) and ila (w/o hamza), so each verb is repeated twice
```

```python
#delete the last character of item in ilas list so that it can be replaced by a * eventually

ilas_without_alifmaqsura = []
for item in all_ilas_final:
    new = item.rstrip(item[-1])
    ilas_without_alifmaqsura.append(new)
ilas_without_alifmaqsura[0:10]
```

```python
# remove anything in all_ilas_final from the all_combos list so that I can add in the abbreviated ones in ilas_without_alifmaqsura

all_combos_without_ilas = []

for combo in all_combos:
    if combo not in all_ilas_final:
        all_combos_without_ilas.append(combo)
len(all_combos_without_ilas)
```

```python
# This list has all the original combos with the abbreviated ilas spelled both ways, but still needs the alas added in
all_combos_withabbrev_ilas = all_combos_without_ilas + ilas_without_alifmaqsura
len(all_combos_withabbrev_ilas)
```

```python
# alas has both ala and alay, but some alays have pronouns attached and all of them need to be trimmed of last char
alas = []
for combo in all_combos:
    if "على" in combo:
        alas.append(combo)
len(alas)
```

```python
alas[0:20]
```

```python
# list of alays only (with and without pronouns on 3lay)
alays = []
for combo in all_combos:
    if "علي" in combo:
        alays.append(combo)
alays
```

```python
# isolating the ones with pronouns attached to prep for removal
alays_prns = []
for item in alays:
    if item[-1] != "ي":
        alays_prns.append(item)
alays_prns
```

```python
# removing the feminine and masculine pronouns
alays_1 = []
for item in alays_prns:
    if item[-1] == "ه":
        newmasc = item.rstrip(item[-1])
        alays_1.append(newmasc)
    if item[-1] =="ا":
        newfem = item.rstrip(item[-1])
        newfem1 = newfem.rstrip(newfem[-1])
        alays_1.append(newfem1)
alays_1
```

```python
# a list of all the alays that had pronouns, now without
alays_noprons = []

for item in alays:
    if item not in alays_prns:
        alays_noprons.append(item)
alays_noprons
```

```python
# adding the pronoun-stripped alays back together with the non-pronoun alays
alays_noprons1 = alays_noprons + alays_1
alays_noprons1
```

```python
# making sure there are no duplicates of the old pronoun-attached alays by making an 3la only list
alas_noalays = []
for combo in alas:
    if "على" in combo:
        alas_noalays.append(combo)
len(alas_noalays)
```

```python
# alas final has all ala and alay with no pronouns attached. Still needs to be trimmed of last character
alas_final = alas_noalays + alays_noprons1
len(alas_final)
```

```python
# trimming off the last character for all ala and alay

alas_without_finalchar = []
for item in alas_final:
    new = item.rstrip(item[-1])
    alas_without_finalchar.append(new)
len(alas_without_finalchar)
```

```python
len(alas_without_finalchar)
```

```python
# removing the old alas and alays that are not trimmed and still have pronouns attached

master_combos = []
for item in all_combos_withabbrev_ilas:
    if item not in alas:
        master_combos.append(item)
master_combos[0:15]
```

```python
# adding alas back into the master list
master_combos_final = master_combos + alas_without_finalchar
print(len(master_combos_final))
print(master_combos_final[0:25])
```

```python
# adding wildcards to all ends of verbs and preps (but not in the middle yet)
wildcards_attached = []

separator = " "

for item in master_combos_final:
    newitem = item.split(separator, 1)
    wildcards_attached.append(newitem)

wildcards_attached[0:15]
```

```python
for sublist in wildcards_attached:
    print(sublist[1])
```

```python
# wildcards_attached1 has all combinations with a wildcard at the end for pronouns, but not in middle 
wildcards_attached1 = []
for sublist in wildcards_attached:
    newitem1 = sublist[0] + "*" + " " + sublist[-1]
    newitem2 = newitem1 + "*"
    wildcards_attached1.append(newitem2)
print(wildcards_attached1[0:10])
```

```python
wildcard_inmiddle = []
for combo in wildcards_attached1:
    newitem = combo.split(separator, 1)
    wildcard_inmiddle.append(newitem)

wildcard_inmiddle[0:15]
```

```python
# wildcards_inmidfinal has all combos with a wildcard in the middle for non-consecutive sequences
wildcards_inmidfinal = []
for sublist in wildcard_inmiddle:
    newitem1 = sublist[0] + " " "*" + " " + sublist[-1]
    wildcards_inmidfinal.append(newitem1)
print(wildcards_inmidfinal[0:10])
```

```python
# this list has all combinations 2x, once without wildcard in middle and once with
final_arabicbatchsearchlist = wildcards_attached1 + wildcards_inmidfinal
print(final_arabicbatchsearchlist[2000:2001])
```
